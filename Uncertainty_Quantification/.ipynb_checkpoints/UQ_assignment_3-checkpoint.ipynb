```python
import numpy as np
import numpy.polynomial.legendre as lg
import scipy as sp
from matplotlib import pyplot as plt
```

```python
a = -1
b =  1
N = 60 # grade = N-1
M = N + 1
f_n_hat = []
variance = 0
z_r = np.random.uniform(-1,1)
f = lambda z_r: np.sin(2*(z_r - 1)**2)
for n in range(1,N+1):
    x, w = np.polynomial.legendre.leggauss(M)
    phi_n = lg.Legendre.basis(n-1)
    c_i = sum(f(x[j]) * w[j] * phi_n(x[j]) for j in range(N))
    g_i = sum(phi_n(x[j])*phi_n(x[j]) * w[j] for j in range(N))
    f_n_hat += [c_i / g_i]
    
    variance += (c_i/g_i)**2 * g_i
    
Pnf = lg.Legendre(f_n_hat).truncate(7)
print(Pnf)
print(variance)
```

```python
t1 = np.arange(a, b, 0.001)

var = np.var(f(t1))

mean = np.average(f(t1))
print(mean)
print(var)

plt.figure()

plt.subplot(211)
plt.plot(t1, Pnf(t1), 'b')
plt.show()
plt.subplot(212)
plt.plot(t1, Pnf(t1), 'b', label='Approximation')
plt.plot(t1, f(t1), 'r', label='Original')
plt.legend()
plt.show()

for i in range(1, 8):
    plt.plot(lg.Legendre.basis(i)(t1))

plt.title('Orthogonal Polynomials')
plt.show()
```
